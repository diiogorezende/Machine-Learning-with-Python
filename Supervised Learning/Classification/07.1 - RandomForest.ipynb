```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, KFold, cross_val_score, RandomizedSearchCV
from sklearn.decomposition import PCA
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('/content/drive/MyDrive/Udemy/ML com Python/Datasets/BRAZIL_CITIES.csv', sep=';', decimal=',')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(['CITY', 'IDHM Ranking 2010', 'IDHM_Renda', 'IDHM_Longevidade', 'IDHM_Educacao', 'LONG', 'LAT', 'GVA_MAIN',
         'REGIAO_TUR', 'MUN_EXPENDIT', 'HOTELS', 'BEDS', 'Pr_Agencies', 'Pu_Agencies', 'Pr_Bank', 'Pu_Bank', 'Pr_Assets',
         'Pu_Assets','UBER','MAC','WAL-MART'], axis=1, inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
df['AREA'] = df['AREA'].str.replace(',', '')
```

```python
# Transformação de variáveis categóricas em numéricas

state_encode = pd.get_dummies(df['STATE']).astype(int)
rural_encode = pd.get_dummies(df['RURAL_URBAN']).astype(int)
categoria_encode = pd.get_dummies(df['CATEGORIA_TUR']).astype(int)

# Remover as variáveis categóricas
df.drop(['STATE', 'RURAL_URBAN', 'CATEGORIA_TUR'], axis=1, inplace=True)

# Concatenar as variáveis OHE no dataset.
df_concat = pd.concat([df, state_encode, rural_encode, categoria_encode], axis=1)
```

```python
df_concat.head()
```

```python
df_concat.shape
```

```python
# Remover os dados faltantes

df_novo = df_concat.dropna()
```

```python
df_novo.shape
```

```python
# Separação das variáveis preditoras e target
X = df_novo.drop('IDHM', axis=1)
y = df_novo['IDHM']
```

```python
# Normalização das variáveis preditoras
normalize = MinMaxScaler()
X_normalised = normalize.fit_transform(X)

# Transformação dos dados usando PCA
pca = PCA(n_components=15)
X_pca = pca.fit_transform(X_normalised)
```

```python
# Verificar a variancia dos componentes
print(f"Variância dos componentes: {pca.explained_variance_ratio_}")
```

```python
# Soma da variancia dos componentes
print(f"Soma da variância dos componentes: {pca.explained_variance_ratio_.sum()}")
```

```python
# Instânciar o modelo
modelo = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

# Instânciar a validação cruzada
kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# Verificação do resultado obtido.
resultado = cross_val_score(modelo, X_pca, y, cv=kfold, n_jobs=-1)
```

```python
resultado.mean()
```

```python
# Avaliação do modelo com diferentes parâmetros

# Valores que vamos testar
min_split = np.array([2, 3, 4, 5])
max_depth = np.array([2, 3, 5, 7, 9, 11, 13])
min_leaf = np.array([3, 4, 5, 6, 7])
val_grid = {'min_samples_split': min_split, 'max_depth': max_depth, 'min_samples_leaf': min_leaf}

# Criar o modelo
modelo2 = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)

# Grids
gridRF = RandomizedSearchCV(estimator=modelo2, param_distributions=val_grid, cv=5, n_iter=50, n_jobs=-1)
gridRF.fit(X_pca, y)
```

```python
# Exibir os melhores parâmetros
gridRF.best_params_
```

```python
# Exibir o melhor score
gridRF.best_score_
```

```python
# Outra verificação
X_tr, X_ts, y_tr, y_ts = train_test_split(X_pca, y, test_size=0.3, random_state=42)

modelo3 = RandomForestRegressor(n_estimators=100,
                                min_samples_split=4,
                                min_samples_leaf=4,
                                max_depth=13,
                                random_state=42)
modelo3.fit(X_tr, y_tr)
```

```python
# Predição
y_pred = modelo3.predict(X_ts)

r_2 = r2_score(y_ts, y_pred)
rmse = np.sqrt(mean_squared_error(y_ts, y_pred))
mae = mean_absolute_error(y_ts, y_pred)

# Exibir as métricas
print(f"R²: {r_2}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
```
