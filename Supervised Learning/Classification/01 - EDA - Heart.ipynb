## Tratamento de valores incoerentes

```python
# Vamos remover os registros onde a presso  zero.
df2 = df.loc[df['RestingBP'] != 0] # Aqui, estamos realizando a seleo de todos diferentes de zero
```

```python
df2.shape
```

```python
df2.describe()
```

## Substituio dos valores zero na varivel colesterol pela mdia.

```python
df2['Cholesterol'].mean()
```

```python
df2["Cholesterol"].replace(0, np.NaN, inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['Cholesterol'].fillna(df2['Cholesterol'].mean(), inplace=True)
```
