## Tratamento de valores incoerentes

```python
# Vamos remover os registros onde a pressão é zero.
df2 = df.loc[df['RestingBP'] != 0] # Aqui, estamos realizando a seleção de todos diferentes de zero
```

```python
df2.shape
```

```python
df2.describe()
```

## Substituição dos valores zero na variável colesterol pela média.

```python
df2['Cholesterol'].mean()
```

```python
df2["Cholesterol"].replace(0, np.NaN, inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2['Cholesterol'].fillna(df2['Cholesterol'].mean(), inplace=True)
```

```python
df2.isnull().sum()
```

```python
df2.describe()
```

```python
sns.histplot(df2, x='Cholesterol', bins=30, color='blue', kde=True, stat='count');
```
